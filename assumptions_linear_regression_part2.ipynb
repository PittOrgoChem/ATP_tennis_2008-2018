### Modeling - Regression

```python
X = house_df[['overallqual', 'grlivarea', 'garagearea', 'firstflrsf', 'totrmsabvgrd', 'exterqual_dummy_Fa', 'exterqual_dummy_Gd', 'exterqual_dummy_TA', 'kitchenqual_dummy_Fa', 'kitchenqual_dummy_Gd', 'kitchenqual_dummy_TA', 'mszoning_dummy_FV', 'mszoning_dummy_RH', 'mszoning_dummy_RL', 'mszoning_dummy_RM']]
Y = house_df['saleprice']
```

```python
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

### Assumption Testing

#### Assumption 1: Linearity of the model in its coefficients

In order to test for this first assumption, the predicted values for the target and the errors in the predictions (residuals) were appended to the working dataframe and the observed values plotted against the predictions

```python
house_df['saleprice_pred'] = results.predict()
```

```python
house_df['residuals'] = results.resid
```

```python
sns.regplot(data = house_df, x = 'saleprice', y = 'saleprice_pred', lowess = True, line_kws = {'color': 'red'})
plt.title('Observed vs. Predicted Values', fontsize = 16)
plt.ylabel('Predicted')
plt.xlabel('Observed');
```

As it can be seen in the plot above, the model can be assumed to be almost linear in its coefficients. In an attempt to further evaluate the assumtion, the relationship between each independent variable used to built the model and the target variable was explored

```python
sns.pairplot(house_df[['saleprice', 'overallqual', 'grlivarea', 'garagearea', 'firstflrsf', 'totrmsabvgrd']]);
```

In the series of plots displayed above, good correlations between each of the numeric predictors and the target is evident. It can be concluded that, although not linear, our model is approximately linear in its coefficients

#### Assumption 2: the error term should be zero on average

```python
print("Mean of the errors in the house price model is: {}".format(np.mean(house_df['residuals'])))
```

The mean of the errors in our predictions clearly approximates zero, therefore, it can be concluded that our working model satisfies this assumption

#### Assumption 3: homoscedasticity

To test for this third assumption, the errors were plotted against the predicted values to evaluate the variance distribution of the errors. As it can be seen in the plot below, predicted values between 0 and 100,000 usually lead to less variance in the errors compared to the values in the middle range (150,000-250,000)

```python
sns.scatterplot(x = house_df['saleprice_pred'], y = house_df['residuals'])
plt.xlabel('Predicted')
plt.ylabel('Residual')
plt.axhline(y = 0)
plt.title('Residual vs. Predicted');
```

To further our analysis, the two statistical test below were performed. Both tests are based on a null hypothesis of the residuals being consistently variant across the predictions. As it can be noted from the results below, both test support rejecting the null hypothesis, therefore suggesting the model is NOT HOMOSCEDASTIC

```python
from scipy.stats import bartlett
from scipy.stats import levene

bart_stats = bartlett(house_df['saleprice_pred'], house_df['residuals'])
lev_stats = levene(house_df['saleprice_pred'], house_df['residuals'])

print("Bartlett test statistic value is {0:3g} and p value is {1:.3g}".format(bart_stats[0], bart_stats[1]))
print("Levene test statistic value is {0:3g} and p value is {1:.3g}".format(lev_stats[0], lev_stats[1]))
```

#### Assumption 4: low multicollinearity

In order to test for this assumption, the Variance Inflation Factors for the predictors used to build the model were calculated. As it can be seen in the table below, only dummy variables created from the same categorical variables report VIF numbers higher than five (5) and sometimes even higher than twenty (20). Since some correlation is expected between dummy variables derived from the same variable, only VIF from numeric variables were considered

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif
```

To further investigate the correlation between the numeric predictors, a correlation matrix was built and displayed below

```python
house_df[['overallqual', 'grlivarea', 'garagearea', 'firstflrsf', 'totrmsabvgrd']].corr()
```

As it can be seen from the matrix above, most of the variables do not exhibit strong correlation with others, however, the variables 'totrmsabvgrd' and 'grlivarea' are more strongly correlated (0.83). This correlation is certainly not negligible and may cause problems in the model

#### Assumption 5: error terms should be uncorrelated with one another

To test for this assumption, a plot of the residuals was constructed and displayed below. The plot shows randomness in the distribution of the errors around zero with the exception of few records associated to larger than usual errors. However, no trend or autocorrelation can be witnessed

```python
sns.lineplot(data = house_df['residuals']);
```

To get a better idea about the presence of any autocorrelation, the autocorrelation factors of the residuals were calculated and plotted as displayed below. In the plot, it can be seen that the errors seem not to be correlated with small autocorrelation factors ranging between 0.06 and -0.06. Based on these results, we should conclude our working model **satisfies this assumption**

```python
from statsmodels.tsa.stattools import acf

acf_data = acf(house_df['residuals'])

plt.plot(acf_data[1:]);
```

#### Assumption 6: features shouldn't be correlated with the errors

In order to explore whether our model satisfies this last assumption, a correlation matrix that includes the 'residuals' and all numeric predictor variables was constructed. Based on the very small correlation between the 'residuals' variable and all other variables explored, it may be concluded that the errors are **NOT CORRELATED** with the features used to build the model

```python
house_df[['residuals', 'overallqual', 'grlivarea', 'garagearea', 'firstflrsf', 'totrmsabvgrd']].corr()
```

In an attempt to confirm the findings shared above, the Pearson coefficients for each numeric predictors against the 'residuals' variable was calculated. The performed test is based in the null hypothesis of the variables not being correlated. From the results below, the obtained p-values suggest that in every cased studied we failed to reject the null hypothesis, therefore confirming that the variables are **NOT CORRELATED**

```python
from scipy.stats.stats import pearsonr

for column in X.columns:
    corr_test = pearsonr(X[column], house_df['residuals'])
    print(f'Variable: {column} --- correlation: {corr_test[0]:.4f}, p-value: {corr_test[1]:.4f}')
```

#### Additional assumption

As a last assumption to be tested, the distribution of the values in the 'residuals' variable was explored using a Q-Q plot and a histogram. Both plots displayed below, strongly suggest a distribution distant from a Normal distribution

```python
rand_nums = np.random.normal(np.mean(house_df['residuals']), np.std(house_df['residuals']), len(house_df['residuals']))

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(np.sort(rand_nums), np.sort(house_df['residuals'])) # we sort the arrays
plt.xlabel("the normally distributed random variable")
plt.ylabel("errors of the model")
plt.title("QQ plot")

plt.subplot(1,2,2)
plt.hist(house_df['residuals'])
plt.xlabel("errors")
plt.title("Histogram of the errors")

plt.tight_layout();
```

To support the finding above on statistical tests, two tests available were performed on the 'residuals' variable and the results shared below. As it can be seen the p-values associated to both tests support the rejection of the null hypothesis, which in both cases is that the distribution of values is normal. Based on this results it can be concluded that the **errors are not normally distributed**

```python
from scipy.stats import jarque_bera
from scipy.stats import normaltest

jb_stats = jarque_bera(house_df['residuals'])
norm_stats = normaltest(house_df['residuals'])

print("Jarque-Bera test statistics is {0} and p value is {1}".format(jb_stats[0], jb_stats[1]))
print("Normality test statistics is {0} and p value is {1}".format(norm_stats[0], norm_stats[1]))
```
